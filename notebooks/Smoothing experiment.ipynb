```python
import nibabel as nb
from nilearn.plotting import plot_glass_brain
from nilearn.image import smooth_img
%matplotlib inline
```

```python
smooth_nii = nb.load("c:/data/neurovault/original/13343.nii.gz")
plot_glass_brain(smooth_nii,
                display_mode='lzry',
                colorbar=True,
                threshold=0,
                plot_abs=False)
```

```python
plot_glass_brain(smooth_img(smooth_nii, 8),
                display_mode='lzry',
                colorbar=True,
                threshold=0,
                plot_abs=False)
```
