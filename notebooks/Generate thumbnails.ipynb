```python
files = glob.glob("D:/data/neurovault/neurovault/original/*.nii.gz")
shuffle(files)
for image in files:
    nii = nb.load(image)
    data = nii.get_data()
    data[np.isnan(data)] = 0
    
    values, counts = np.unique(data[data != 0], return_counts=True)
    is_clipped = counts.max() > 10 and ((values[np.argmax(counts)] == data.max()) or (values[np.argmax(counts)] == data.min()))
    is_segmented = len(values) < 500
    
    values, counts = np.unique(data, return_counts=True)
    is_offzero = values[np.argmax(counts)] != 0
    
    is_toolarge = (np.abs(data) > 100).any()
    
    is_toosmall = (np.abs(data) < 1).all()
    
    no_negative_values = np.min(data) >=0
    
    no_positive_values = np.max(data) <=0
    
    smoothness_file = image.replace(".nii.gz", "_smoothness.txt")
    no_smoothness_file = not os.path.exists(smoothness_file)
    if no_smoothness_file:
        is_toocoarse = -1
        is_smoothness_aniso = -1
    else:
        smoothness_str = open(smoothness_file).read()
        smoothness_arr = np.array(smoothness_str.split()).astype(np.float)
        mean_smoothness = smoothness_arr.mean()
        std_smoothness = smoothness_arr.std()
        is_toocoarse = mean_smoothness < 7.0
        is_smoothness_aniso = std_smoothness > 4.0
        
    image_file = image.replace("original", "original/figures").replace(".nii.gz", ".png")
    no_image_file = not os.path.exists(image_file)
    
    if is_toolarge or is_toosmall or no_negative_values or no_positive_values or is_segmented or is_clipped or is_offzero or no_smoothness_file or is_toocoarse or is_smoothness_aniso or no_image_file:
        print(image)
        print("is_clipped = %g"%is_clipped)
        print("is_segmented = %g"%is_segmented)
        print("is_offzero = %g"%is_offzero)
        print("is_toolarge = %g"%is_toolarge)
        print("no_negative_values = %g"%no_negative_values)
        print("no_positive_values = %g"%no_positive_values)
        print("is_toosmall = %g"%is_toosmall)
        print("no_smoothness_file = %g"%no_smoothness_file)
        print("is_toocoarse = %g"%is_toocoarse)
        print("is_smoothness_aniso = %g"%is_smoothness_aniso)
        print("no_image_file = %g"%no_image_file)
        
        shutil.move(image, image.replace("original", "rejected"))
        if not no_image_file:
            shutil.move(image_file, image_file.replace("original", "rejected"))
        if not no_smoothness_file:
            shutil.move(smoothness_file, smoothness_file.replace("original", "rejected"))
            
```

```python
def plot(path):
    fig = plt.figure(figsize=(6,2))
    try: 
        plot_glass_brain(path,
                display_mode='lzry',
                colorbar=True,
                threshold=0,
                plot_abs=False,
                        figure=fig)
    except ValueError:
        print("failed to parse %s"%path)
    else:
        plt.savefig(path.replace("original", "original/figures").replace(".nii.gz", ".png"))
    finally:
        plt.close(fig)
    return 0
```

```python
f = open("D:/data/neurovault/neurovault/rejection_log.txt")
path = ""
is_clipped = False
all_zero = False
for line in f:
    if line.startswith("D:"):
        if is_clipped and all_zero:
            src_path = path.replace("original", "rejected/figures").replace(".nii.gz", ".png").replace("\n", "")
            dst_path = src_path.replace("rejected", "clipped")
            try:
                shutil.move(src_path, dst_path)
            except:
                print("failed to move %s"%src_path)
        path = line
        is_clipped = False
        all_zero = True
    elif line.startswith("is_clipped = 1"):
        is_clipped = True
    elif line.endswith("= 1\n"):
        all_zero = False
```

```python
for file in glob.glob("D:/data/neurovault/neurovault/original/figures/*.png"):
    nii_file_orig = file.replace("/figures", "").replace(".png", ".nii.gz")
    nii_file_rejected = file.replace("/figures", "").replace(".png", ".nii.gz").replace("original", "rejected")
    smoothness_file_rejected = nii_file_rejected.replace(".nii.gz", "_smoothness.txt")
    smoothness_file_orig = nii_file_orig.replace(".nii.gz", "_smoothness.txt")
    if not os.path.exists(nii_file_orig):
        shutil.move(nii_file_rejected, nii_file_orig)
        shutil.move(smoothness_file_rejected, smoothness_file_orig)
```

```python
for file in glob.glob("D:/data/neurovault/neurovault/rejected/figures/*.png"):
    nii_file_rejected = file.replace("/figures", "").replace(".png", ".nii.gz")
    nii_file_orig = file.replace("/figures", "").replace(".png", ".nii.gz").replace("rejected", "vetted")
    smoothness_file_rejected = nii_file_rejected.replace(".nii.gz", "_smoothness.txt")
    smoothness_file_orig = nii_file_orig.replace(".nii.gz", "_smoothness.txt")
    if not os.path.exists(nii_file_rejected):
        #print(nii_file_rejected)
        shutil.move(nii_file_orig, nii_file_rejected)
        shutil.move(smoothness_file_orig, smoothness_file_rejected)
```
