```python
from glob import glob
```

```python
files_to_be_moved = []
for f in glob("D:/data/neurovault/neurovault/vetted/eval/*.nii.gz"):
    for id in ids:
        if id in f:
            files_to_be_moved.append(f)
```

```python
files_to_be_moved = set(files_to_be_moved)
import shutil
```
