```python
from pyneurovault import api
```

```python
images = api.get_images()
```

```python
orig_images = images
```

```python
for column in orig_images.columns:
    print(column)
```

```python
filtered = orig_images.query('is_thresholded==False')
len(filtered)
```

```python
filtered = filtered.query('not_mni==False')
len(filtered)
```

```python
set(filtered.analysis_level)
```

```python
filtered = filtered.query("analysis_level!='meta-analysis'")
len(filtered)
```

```python
set(filtered.modality)
```

```python
filtered = filtered.query("modality!='Structural MRI'")
filtered = filtered.query("modality!='PET FDG'")
filtered = filtered.query("modality!='PET other'")
filtered = filtered.query("modality!='MEG'")
filtered = filtered.query("modality!='Diffusion MRI'")
filtered = filtered.query("modality!='EEG'")
filtered = filtered.query("modality!='fMRI-CBF'")
len(filtered)
```

```python
set(filtered.cognitive_paradigm_cogatlas)
```

```python
filtered = filtered.query("cognitive_paradigm_cogatlas!='None / Other'")
filtered = filtered.query("cognitive_paradigm_cogatlas!='rest eyes closed'")
filtered = filtered.query("cognitive_paradigm_cogatlas!='rest eyes open'")
len(filtered)
```

```python
len(set(filtered.collection_id))
```

```python
filtered = filtered.query("collection_id not in [2606, 2572, 2576, 2575, 2574, 2579]") #CNP group level maps
len(filtered)
```

```python
set(filtered.image_type)
```

```python
len(filtered.query("analysis_level!='single-subject'"))
```
